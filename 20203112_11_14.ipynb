```python
from google.colab.patches import cv2_imshow
import cv2
from google.colab import drive
drive.mount("/content/drive/")
import numpy as np
```

```python
imgp='/content/drive/MyDrive/down.jpg'
img=cv2.imread(imgp)
cv2_imshow(img)
cv2.imwrite('/content/drive/MyDrive/img.jpg',img)
res2=cv2.resize(img,dsize=(227*2,180*2))
cv2_imshow(res2)
cv2.imwrite('/content/drive/MyDrive/res2.jpg',res2)
res5=cv2.resize(img,dsize=(113,90))
cv2_imshow(res5)
cv2.imwrite('/content/drive/MyDrive/res5.jpg',res5)
```

```python
(b, g, r) = img[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

```python
wing = img[10:200, 20:150]
cv2_imshow(wing)
cv2.imwrite('/content/drive/MyDrive/wing.jpg',wing)
```
